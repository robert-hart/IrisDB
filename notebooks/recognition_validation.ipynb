```python
import os
import shutil
import numpy as np
import pandas as pd
import seaborn as sns
import cv2
from matplotlib import pyplot as plt

path = '/Users/roberthart/Documents/GitHub/IrisDB'

path_data_imposter = f'{path}/data/datasets/recognition-verification/verification-intra-imposter.parquet'

path_data_inter_rotations = f'{path}/data/datasets/recognition-verification/verification-inter-rotations.parquet'
```

```python
HD_real = np.load(f'{path}/real_dist_combined/HD_NP.npy')
HD_imposter = np.load(f'{path}/imposter_dist/4_sorted_by_hd.npy')
```

```python
plt.figure(figsize=(7.16, 5.37))

sns.set_theme(style="ticks")

colors = sns.color_palette("colorblind")
plt.rcParams['font.family'] = 'Times New Roman'

sns.histplot(HD_real, bins=30, kde=True, color=colors[0], alpha=0.5, label='Authentic Comparisons', stat="density")
sns.histplot(HD_imposter, bins=30, kde=True, color=colors[3], alpha=0.5, label='Imposter Comparisons', stat="density")
plt.title('Authentic Comparisons vs Imposter Comparisons', fontsize=16)

plt.legend(loc='best', fontsize=10)
plt.xlabel('Hamming Distance', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=10)  # Adjust 'labelsize' as needed

plt.savefig('true+imposter_training.png', dpi=600)
# Show the plot with Seaborn's enhanced visuals
plt.show()
```
