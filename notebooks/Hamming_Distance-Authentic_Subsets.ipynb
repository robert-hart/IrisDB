```python
import os
import shutil
import numpy as np
import pandas as pd
import seaborn as sns
import cv2
from matplotlib import pyplot as plt
import matplotlib.ticker as mticker
```

```python

path = '' #path to directory containing data
colors = sns.color_palette("colorblind")
colors_hex = colors.as_hex()

labelpad_distance = 10

plt.rcParams['font.family'] = 'Times New Roman'

holepunched = np.load(f'{path}/HD_holepunched.npy') #loads hamming distances calculated from 'authentic' comparisons; 'holepunched' subdivision only.
plt.figure(figsize=(7.16, 5.37))
plt.title("Distribution of 'Hole Punched' Authentic Comparisons", fontsize=16)
sns.histplot(holepunched, bins=20, kde=False, color=colors[0], alpha=0.5,label='True Match Distribution Holepunched', stat="density")
plt.xlabel('Hamming Distance', fontsize=14, labelpad=labelpad_distance)
plt.ylabel('Density', fontsize=14, labelpad=labelpad_distance)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xlim(0, 0.5)
plt.show()
```

```python
plt.clf()
rotated = np.load(f'{path}/HD_rotated.npy') #loads hamming distances calculated from 'authentic' comparisons; 'rotated' subdivision only.
plt.figure(figsize=(7.16, 5.37))
plt.title("Distribution of 'Rotated' Authentic Comparisons", fontsize=16)
sns.histplot(rotated, bins=20, kde=False, color=colors[0], alpha=0.5,label='True Match Distribution Holepunched', stat="density")
plt.xlabel('Hamming Distance', fontsize=14, labelpad=labelpad_distance)
plt.ylabel('Density', fontsize=14, labelpad=labelpad_distance)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xlim(0, 0.5)
plt.show()
```

```python
plt.clf()

data = np.concatenate([rotated, holepunched]) #combines hamming distances from both subdivisions
plt.figure(figsize=(7.16, 5.37))
plt.title('Distribution of All Authentic Comparisons', fontsize=16)
sns.histplot(data, bins=20, kde=False, color=colors[0], alpha=0.5,label='True Match Distribution Holepunched', stat="density")
plt.xlabel('Hamming Distance', fontsize=14, labelpad=labelpad_distance)
plt.ylabel('Density', fontsize=14, labelpad=labelpad_distance)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xlim(0, 0.5)
plt.show()
```
