```python
#import required libraries
import os
import shutil
import numpy as np
import pandas as pd
import seaborn as sns
import cv2
from matplotlib import pyplot as plt
```

```python
path = '' #insert path to root data directory

colors = sns.color_palette("colorblind")
colors_hex = colors.as_hex()

plt.rcParams['font.family'] = 'Times New Roman'

inter = np.load(f'{path}/inter_dataset_distances.npy', allow_pickle=True) #load numpy array of inter dataset hamming distances from .npy file called 'inter_dataset_distances.npy'
inter = inter.flatten()
plt.figure(figsize=(7.16, 5.37))
 
sns.histplot(inter, bins=30, kde=False, color=colors[0], alpha=0.5, stat="density")
plt.title('Lowest Inter-Dataset Comparisons', fontsize=16)
plt.xlabel('Euclidean Distance', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```

```python
plt.clf()
intra = np.load(f'{path}/intra_dataset_distances.npy', allow_pickle=True) ##load numpy array of intra dataset hamming distances from .npy file called 'intra_dataset_distances.npy'
intra = intra.flatten()
plt.figure(figsize=(7.16, 5.37))
 
sns.histplot(intra, bins=30, kde=False, color=colors[0], alpha=0.5, stat="density")
plt.title('Intra-Training Set Comparisons', fontsize=16)
plt.xlabel('Euclidean Distance', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```
