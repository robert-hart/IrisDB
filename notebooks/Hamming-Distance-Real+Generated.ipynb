```python
#Load required libraries

import os
import shutil
import numpy as np
import pandas as pd
import seaborn as sns
import cv2
from matplotlib import pyplot as plt
import matplotlib.ticker as mticker
```

```python

path = '' #path to data directory
colors = sns.color_palette("colorblind")

training = np.load(f'{path}/HD_training.npy', allow_pickle=True) #loads hamming distances calculated by comparing training images to each other.
samples = np.load(f'{path}/HD_samples.npy', allow_pickle=True) ##loads hamming distances calculated by comparing sampled images to each other.
plt.figure(figsize=(7.16, 5.37))

sns.kdeplot(training, color=colors[0], label="Real Irises", linewidth=2.5)
sns.kdeplot(samples, color=colors[3], label="Generated Irises", linewidth=2.5)

plt.legend(loc='best', fontsize=10)

plt.rcParams['font.family'] = 'Times New Roman'
plt.title('Real Irises vs. Generated Irises', fontsize=16)
plt.xlabel('Hamming Distance', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xlim(0.3, 0.5)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```
